## Crop image

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('images/cat.jpg')
plt.imshow(image[...,::-1])
```

## Let's go crop the photo.

```python
# height * width 
image.shape[:2]
```

```python
# we can use tuple unpacking
height, width = image.shape[:2]

# use indexing to crop
cropped = image[115:390, 135:430]

plt.imshow(cropped[...,::-1])
```

```python
#show with open cv
cv2.imshow('Original', image)
cv2.imshow('Cropped', cropped)

cv2.waitKey()
cv2.destroyAllWindows()
```

### Using selectROI()

```python
image = cv2.imread('images/cat.jpg')

rect = cv2.selectROI(image) 

x, y, w, h = rect

cropped_img = image[y:y+h, x:x+w]

cv2.imshow('Cropped Image', cropped_img)
cv2.imwrite('new.jpg',cropped_img)

cv2.waitKey()
cv2.destroyAllWindows()
```
