### HSV

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

### Color space conversion

```python
img = cv2.imread('images/bird.jpg')
hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV) # this flog number is 40
cv2.imshow('HSV image', hsv)
cv2.imshow('Hue',hsv[:,:,0])
cv2.imshow('Saturation', hsv[:,:,1])
cv2.imshow('Value',hsv[:,:,2])
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# show pic in matplotlib
plt.figure(figsize=[20,5])
plt.subplot(131);plt.imshow(hsv[:,:,0], cmap='gray');plt.title('Hue')
plt.subplot(132);plt.imshow(hsv[:,:,1], cmap='gray');plt.title('Saturation')
plt.subplot(133);plt.imshow(hsv[:,:,2], cmap='gray');plt.title('Value')
```

### Applying filter like instagram

```python
# change Saturation
saturated_image = cv2.merge([hsv[:, :, 0], cv2.add(hsv[:, :, 1],50), hsv[:, :, 2]])
saturated_image = cv2.cvtColor(saturated_image, cv2.COLOR_HSV2BGR)

plt.figure(figsize=[12,5])

plt.subplot(121)
plt.imshow(saturated_image[...,::-1])
plt.title('saturated image')

plt.subplot(122)
plt.imshow(img[...,::-1])
plt.title('original image')
                           
```

```python
# change Value
# This is haw you can brighten dark photos
brighter_image = cv2.merge([hsv[:, :, 0], hsv[:, :, 1], cv2.add(hsv[:, :, 2], 50)])
brighter_image = cv2.cvtColor(brighter_image, cv2.COLOR_HSV2BGR)

plt.figure(figsize=[12,5])

plt.subplot(121)
plt.imshow(brighter_image[...,::-1])
plt.title('brighter image')

plt.subplot(122)
plt.imshow(img[...,::-1])
plt.title('original image')
```
