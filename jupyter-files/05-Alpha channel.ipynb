```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/transparency.png', cv2.IMREAD_UNCHANGED) # Dont read as BGR
img.shape #The fourth channel is called the alpha channel.
```

```python
# Show channels
bgr = img[...,0:3]
alpha = img[...,-1]
rgb = cv2.cvtColor(bgr,cv2.COLOR_BGR2RGB)
plt.imshow(rgb);plt.title('RGB')
```

```python
plt.imshow(alpha, cmap = 'gray');plt.title('Alpha')
```
