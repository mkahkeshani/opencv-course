```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# square
square = np.zeros((300,300),np.uint8)
cv2.rectangle(square,(50,50), (250,250), 255, -1)

# ellipse
ellipse = np.zeros((300,300), np.uint8)
cv2.ellipse(ellipse,(150,150),(150,150),30,0,180,255,-1)

plt.subplot(121);plt.imshow(square,cmap='gray');plt.title('Square')
plt.subplot(122);plt.imshow(ellipse,cmap='gray');plt.title('Ellipse')
```

## Lets bitwise

```python
# show only where they intersect 
# The dimensions of both images must be the same in height and width
bitwise_and = cv2.bitwise_and(square,ellipse)

# show where either square or ellipse is 
bitwise_or = cv2.bitwise_or(square,ellipse)

# show where either exist   
bitwise_xor = cv2.bitwise_xor(square,ellipse)

# show everythin that is not in the square  
bitwise_not_sq = cv2.bitwise_not(square)

# Because they are single-channel, they are displayed in gray
plt.figure(figsize= [12,6])
plt.subplot(242);plt.imshow(square, cmap='gray');plt.title('Square')
plt.subplot(243);plt.imshow(ellipse, cmap='gray');plt.title('Ellipse')
plt.subplot(245);plt.imshow(bitwise_and, cmap='gray');plt.title('AND')
plt.subplot(246);plt.imshow(bitwise_or, cmap='gray');plt.title('OR')
plt.subplot(247);plt.imshow(bitwise_xor, cmap='gray');plt.title('Xor')
plt.subplot(248);plt.imshow(bitwise_not_sq, cmap='gray');plt.title('Not')
```

```python
img = cv2.imread('images/cat.jpg')
cropped = img[100:400, 100:500]

circle = np.zeros((300,400,3), np.uint8)
cv2.circle(circle,(200,180),200,(255,255,255),-1)

output =cv2.bitwise_and(cropped, circle)
plt.imshow(output[...,::-1])
```

### Make an alpha channel

```python
png_img = np.zeros((300,400,4), np.uint8)
png_img[...,0:3] = cropped
circle = np.zeros((300,400), np.uint8)
cv2.circle(circle,(200,180),200,(255,255,255),-1)
png_img[...,3] = circle
cv2.imwrite('cat.png', png_img)
```
