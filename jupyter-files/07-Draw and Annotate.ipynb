```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create a  black image
color_image = np.zeros((500,500,3), np.uint8)

# create a gray image
gray_image = np.zeros((500,500), np.uint8)

# create a png image
png_image = np.zeros((500,500,4), np.uint8)

plt.figure(figsize=[12,3])
plt.subplot(131);plt.imshow(color_image, cmap= 'gray');plt.title('color')
plt.subplot(132);plt.imshow(gray_image, cmap= 'gray');plt.title('Gray')
plt.subplot(133);plt.imshow(png_image);plt.title('Png')
```

### Draw a Line
#### cv2.line(image, starting cordinates, ending cordinates, color, thickness)

```python
img = np.zeros((500,500,3), np.uint8)
cv2.line(img, (0,0), (499,499),(255,127,0), 5)
cv2.imshow('Blue', img)
plt.imshow(img[...,::-1])
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Draw a Rectangle
#### cv2.rectangle(image, starting vertex, opposite vertex, color, thickness)

```python
img = np.zeros((500,500,3), np.uint8)

# If the thickness number is negative, the rectangle will be solid.
cv2.rectangle(img,(100,100),(300,400),(0,125,255),5)
cv2.imshow('Orange', img)
plt.imshow(img[...,::-1])
cv2.waitKey(2000)
cv2.destroyAllWindows()
```

### Draw a Rectangle on an image

```python
img = cv2.imread('images/dog.jpg')
cv2.rectangle(img,(98,33),(160,103),(0,125,255),1)
cv2.imshow('Detect', img)
plt.imshow(img[...,::-1])
cv2.waitKey(2000)
cv2.destroyAllWindows()
```

### Draw a Cricle
#### cv2.circle(image, center, radius, color, thickness)

```python
img = np.zeros((500,500,3), np.uint8)
cv2.circle(img,(350,350), 100,(0,255,0), 5)
cv2.imshow('Circle', img)
plt.imshow(img[...,::-1])
cv2.waitKey(2000)
cv2.destroyAllWindows()
```

### Draw an ellipse
#### cv2.ellipse(image, center, radius, angle, startAngle, endAngle color, thickness)

```python
img = np.zeros((500,500,3), np.uint8)
cv2.ellipse(img,(250,250),(250,150),30,0,360,(177,140,65), 5)
cv2.imshow('Ellipse', img)
plt.imshow(img[...,::-1])
cv2.waitKey(2000)
cv2.destroyAllWindows()
```
