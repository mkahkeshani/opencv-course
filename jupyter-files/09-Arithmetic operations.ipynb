```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('images/bird.jpg')

# create a new matrix to add a value to all channels
# it create a matrix like our image * 100
new_matrix = np.ones(img.shape, dtype = 'uint8') * 100

# add
added = cv2.add(img, new_matrix)

# subtract
sub = cv2.subtract(img, new_matrix)

# check new images
plt.figure(figsize=[15,5])
plt.subplot(131);plt.imshow(img[...,::-1]);plt.title('Original')
plt.subplot(132);plt.imshow(added[...,::-1]);plt.title('Added')
plt.subplot(133);plt.imshow(sub[...,::-1]);plt.title('Subtract')
```
