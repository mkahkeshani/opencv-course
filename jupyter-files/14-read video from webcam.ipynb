```python
import cv2
import matplotlib.pyplot as plt
```

```python
cap = cv2.VideoCapture(0)

while True:
    ret,frame = cap.read()
    cv2.imshow('Webcam', frame)
    if cv2.waitKey(1) & 0xFF == 27:
        break
cap.release()
cv2.destroyAllWindows()
```

### flip video

```python
cap = cv2.VideoCapture(0)

while True:
    ret,frame = cap.read()
    flipped = cv2.flip(frame,1)
    cv2.imshow('webcam', flipped)
    if cv2.waitKey(1) & 0xFF == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
