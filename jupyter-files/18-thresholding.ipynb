```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

### threshold with numpy

```python
img = cv2.imread('images/gradient.jpg')

new = np.copy(img)
new[new < 125] = 0
new[new >= 125] = 255

plt.figure(figsize=[10,5])
plt.subplot(121);plt.imshow(img, cmap='gray');plt.title('Original')
plt.subplot(122);plt.imshow(new, cmap='gray');plt.title('thresh')
```

### threshold with open cv
#### threshold(src,thresh,maxVal,type[   ,dst])

```python
img = cv2.imread('images/plates.jpg')

# our thrsh is 125
# in binary val below 125 goes to 0 and above it goes to 255
ret0,thresh0 = cv2.threshold(img, 125, 255, cv2.THRESH_BINARY)

# below 125 goes to 255 above it goes to 0
ret1,thresh1 = cv2.threshold(img, 125, 255, cv2.THRESH_BINARY_INV)

# above 125 are trunscated to 125
ret2,thresh2 = cv2.threshold(img, 125, 255, cv2.THRESH_TRUNC)

# below 125 goes to 0 and above it unchanged 
ret3,thresh3 = cv2.threshold(img, 125, 255, cv2.THRESH_TOZERO)

# reverse
ret4,thresh4 = cv2.threshold(img, 125, 255, cv2.THRESH_TOZERO_INV)

plt.figure(figsize=[9,6])
plt.subplot(231);plt.imshow(thresh0, cmap = 'gray');plt.title('thresh binary')
plt.subplot(232);plt.imshow(thresh1, cmap = 'gray');plt.title('thresh binary inverse')
plt.subplot(233);plt.imshow(thresh2, cmap = 'gray');plt.title('thresh trunc')
plt.subplot(234);plt.imshow(thresh2, cmap = 'gray');plt.title('thresh  to zero')
plt.subplot(235);plt.imshow(thresh4, cmap = 'gray');plt.title('thresh  to zero inverse')
plt.subplot(236);plt.imshow(img, cmap = 'gray');plt.title('original')
```
