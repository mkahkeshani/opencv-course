### Note: If you encounter anything strange while performing the operation, it is better to know that our assumption is always that our background is completely black and the image on which the operation is performed is white.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/opencv_inv.png', 0)

kernel = np.ones((5,5), np.uint8) # this is our structural element 

# iterations means how many times the operation is performed
erod = cv2.erode(img, kernel, iterations = 1) 
dilat = cv2.dilate(img, kernel, iterations = 1)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

plt.figure(figsize= [10,8])
plt.subplot(321);plt.imshow(img, cmap='gray');plt.title('Orginal')
plt.subplot(323);plt.imshow(erod, cmap='gray');plt.title('Erosion')
plt.subplot(324);plt.imshow(dilat, cmap='gray');plt.title('Dilation')
plt.subplot(325);plt.imshow(opening, cmap='gray');plt.title('Opening')
plt.subplot(326);plt.imshow(closing, cmap='gray');plt.title('closing')
```

### Pre-built kernels in opencv
### cv2.getStructuringElement(shape, ksize)

```python
img = cv2.imread('images/opencv_inv.png', 0)

kernel = cv2.getStructuringElement(cv2.MORPH_CROSS, (5,5),(2,2)) # last arg is Origin 

# iterations means how many times the operation is performed
erod = cv2.erode(img, kernel, iterations = 1) 
dilat = cv2.dilate(img, kernel, iterations = 1)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

plt.figure(figsize= [10,8])
plt.subplot(321);plt.imshow(img, cmap='gray');plt.title('Orginal')
plt.subplot(323);plt.imshow(erod, cmap='gray');plt.title('Erosion')
plt.subplot(324);plt.imshow(dilat, cmap='gray');plt.title('Dilation')
plt.subplot(325);plt.imshow(opening, cmap='gray');plt.title('Opening')
plt.subplot(326);plt.imshow(closing, cmap='gray');plt.title('closing')
```
