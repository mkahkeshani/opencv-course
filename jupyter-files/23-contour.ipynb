```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Read image
img = cv2.imread('images/shapes.jpg')
plt.imshow(img[...,::-1])
```

```python
# convert to gray
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray,cmap='gray')
```

```python
# Threshold to binary
ret,binary = cv2.threshold(img_gray,0,255,cv2.THRESH_BINARY + cv2.THRESH_OTSU)
plt.imshow(binary, cmap='gray')
```

```python
# Inverse binary
inverted = cv2.bitwise_not(binary)
plt.imshow(inverted, cmap='gray')
```

```python
# Find contours and draw
contours, hierarchy = cv2.findContours(inverted, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
color_with_contours = img.copy()
cv2.drawContours(color_with_contours, contours, -1, (0, 255, 0), 2)
plt.imshow(color_with_contours[...,::-1])
```

```python
for contour in contours:
    print(contour.shape)
```
