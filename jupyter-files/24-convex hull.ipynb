```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random
```

### find convex hull by points

```python
random_points = np.random.randint(100,500,20)
random_points = random_points.reshape(10,2)
img = np.zeros((512,512,3),np.uint8)
for i in random_points:
    cv2.circle(img,(i[0],i[1]),5,(255,0,0),-1)
plt.imshow(img[...,::-1])
```

```python
hull = cv2.convexHull(random_points)
hull.shape
```

```python
cv2.drawContours(img,[hull],-1,(0,0,255),4)
plt.imshow(img[...,::-1])
```

### convex in shapes

```python
img = cv2.imread('images/shapes.jpg')
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret,binary = cv2.threshold(img_gray,0,255,cv2.THRESH_BINARY + cv2.THRESH_OTSU)
inverted = cv2.bitwise_not(binary)
contours, hierarchy = cv2.findContours(inverted, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# find convex hull for each contour
hull_list = []
for i in range(len(contours)):
    hull = cv2.convexHull(contours[i])
    hull_list.append(hull)
    
# draw convex and conturs
cv2.drawContours(img, contours,-1,(0,255,0), 1)
cv2.drawContours(img, hull_list, -1,(0,0,255),1)

plt.imshow(img[...,::-1])
```
