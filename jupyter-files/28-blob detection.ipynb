```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# read image
img = cv2.imread('images/blob.jpg')
if img is None:
    print('erorr')

#detect blobd
detector = cv2.SimpleBlobDetector_create()
keypoints = detector.detect(img)

# draw blobs
result1 = cv2.drawKeypoints(img, keypoints, np.array([]),(0,0,255))
result2 = cv2.drawKeypoints(img, keypoints, np.array([]),(0,0,255),cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# show with matplotlib
plt.figure(figsize = [15,15])
plt.subplot(131);plt.imshow(img[...,::-1]);plt.title('Original')
plt.subplot(132);plt.imshow(result1[...,::-1]);plt.title('Result1')
plt.subplot(133);plt.imshow(result2[...,::-1]);plt.title('Result2')
```

### Filter by Area

```python
# read image
img = cv2.imread('images/blob_size.jpg')
if img is None:
    print('erorr')
    
# prameters
params = cv2.SimpleBlobDetector_Params()

# filter by area
params.filterByArea = True
params.minArea = 5000
params.maxArea = 20000

#detect blobd
detector = cv2.SimpleBlobDetector_create(params)
keypoints = detector.detect(img)

# draw blobs

result = cv2.drawKeypoints(img, keypoints, np.array([]),(0,0,255),cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# show with matplotlib

plt.imshow(result[...,::-1]);plt.title('result')
```
