```python
import cv2
import matplotlib.pyplot as plt
```

```python
face =cv2.CascadeClassifier()
face.load('haarcascade_frontalface_alt.xml')

src = cv2.imread('images/pic.png')
if src is None:
    print('Error')
    
gray_src = cv2.cvtColor(src,cv2.COLOR_BGR2GRAY)
gray_src = cv2.equalizeHist(gray_src)

result = face.detectMultiScale(gray_src)

for (x,y,w,h) in result:
    cv2.rectangle(src,(x,y),(x+w, y+h), (0,255,0), 3)
plt.imshow(src[...,::-1])
```
