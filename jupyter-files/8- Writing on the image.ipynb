```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img  = np.zeros((500,500,3), np.uint8)
cv2.putText(img,'Hello World',(70,250),cv2.FONT_HERSHEY_COMPLEX,2,(148,250,98) ,3)
cv2.imshow('Hello', img)
plt.imshow(img[...,::-1])
cv2.waitKey(2000)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('images/cat.jpg')
plt.imshow(img[...,::-1])
```

```python
img = cv2.imread('images/cat.jpg')
cv2.rectangle(img,(138,117),(425,395),(0,255,0),2)
plt.imshow(img[...,::-1])
```

```python
img = cv2.imread('images/cat.jpg')
cv2.rectangle(img,(138,117),(425,395),(0,255,0),2)
cv2.putText(img,'cat!',(145,110),cv2.FONT_ITALIC,0.5,(255,255,255),1, cv2.LINE_AA)
plt.imshow(img[...,::-1])
```
